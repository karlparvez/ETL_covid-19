```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import json
import random
from datetime import datetime #for handling dates
```

```python
confirmed_df = pd.read_csv('time_series_covid_19_confirmed.csv')

confirmed_df.columns = [str(col) + '_C' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in confirmed_df.columns]



confirmed_df
```

```python
deaths_df = pd.read_csv('time_series_covid_19_deaths.csv')

deaths_df.columns = [str(col) + '_D' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in deaths_df.columns]


deaths_df
```

```python
recovered_df = pd.read_csv('time_series_covid_19_recovered.csv')


recovered_df.columns = [str(col) + '_R' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in recovered_df.columns]

recovered_df
```

```python
merged_df1 = pd.merge(confirmed_df, deaths_df, on=["Province/State", "Country/Region", "Lat", "Long"])
merged_df2 = pd.merge(merged_df1, recovered_df, on=["Province/State", "Country/Region", "Lat", "Long"])
merged_df1
```
