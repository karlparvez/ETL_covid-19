```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import json
import random
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
url = 'https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
#print(soup.prettify())
```

```python
scraped = soup.find_all('a', class_='resource-url-analytics')
confirmed = scraped[0]['href']
deaths = scraped[1]['href']
recovered = scraped[2]['href']
scraped
```

```python
confirmed_df = pd.read_csv(confirmed)
deaths_df = pd.read_csv(deaths)
recovered_df = pd.read_csv(recovered)
```

```python
confirmed_df.columns = [str(col) + '_C' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in confirmed_df.columns]

confirmed_df
```

```python
deaths_df.columns = [str(col) + '_D' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in deaths_df.columns]

deaths_df
```

```python
recovered_df.columns = [str(col) + '_R' if col != 'Long' and col != 'Lat' and col != 'Province/State' 
                        and col != 'Country/Region'
                        else col for col in recovered_df.columns]

recovered_df
```

```python
merged_df1 = pd.merge(confirmed_df, deaths_df, on=["Province/State", "Country/Region", "Lat", "Long"])

merged_df1
```

```python
merged_df2 = pd.merge(merged_df1, recovered_df, on=["Province/State", "Country/Region", "Lat", "Long"])

merged_df2
```
